```python
import psycopg2
```

```python
conn = psycopg2.connect(dbname='imdb',user='postgres',password='postgres')
```

```python
curs = conn.cursor()
```

```python
curs.execute("""SELECT table_name FROM information_schema.tables
       WHERE table_schema = 'public'""")
table_list = []
for i in curs.fetchall():
    table_list.append(i[0])
```

```python
import numpy as np
table_list = np.sort(table_list)
```

```python
table_dict = {}
print(type(table_dict))
for i, table in enumerate(table_list):
    table_dict[table] = i
```

```python
table_dict
```

```python
import pickle
outfile = open('table_dict_pickle','wb')
pickle.dump(table_dict, outfile)
```

```python
outfile.close()
```

```python
for table in table_list:
    print(table)
```

```python
col_list = []
for table in table_list:
    curs.execute("SELECT * FROM " + table + " LIMIT 0")
    colnames = [desc[0] for desc in curs.description]
    for col in colnames:
        col_list.append(table+'.'+col)
```

```python
col_list = np.sort(col_list)
```

```python
col_dict = {}
for i, col in enumerate(col_list):
    col_dict[col] = i
```

```python
table_dict
```

```python
col_dict
```

```python
outfile = open('column_dict_pickle', 'wb')
pickle.dump(col_dict, outfile)
```

```python
outfile.close()
```

```python
min_list = []
max_list = []
for table in table_list:
    curs.execute("SELECT * FROM "+table)
    colnames = [desc[0] for desc in curs.description]
    print(colnames)
    for col in colnames:
        curs.execute("SELECT MIN("+col+") FROM "+table)
        tmp = curs.fetchall()
        min_list.append(tmp[0])
        curs.execute("SELECT MAX("+col+") FROM "+table)
        tmp = curs.fetchall() 
        max_list.append(tmp[0])
```

```python
min_list
```

```python
max_list
```

```python
min_list_dict = {}
for i, item in enumerate(min_list):
    min_list_dict[i] = item[0]
```

```python
min_list_dict
```

```python
max_list_dict = {}
for i, item in enumerate(max_list):
    max_list_dict[i] = item[0]
```

```python
max_list_dict
```

```python
import pickle
```

```python
outfile = open('min_dict_pickle', 'wb')
pickle.dump(min_list_dict, outfile)
outfile.close()
```

```python
outfile = open('max_dict_pickle', 'wb')
pickle.dump(max_list_dict, outfile)
outfile.close()
```
